```python
# Import Libraries needed
import pandas as pd
import numpy as np
```

```python
# Open csv file and read the first 5 lines

df = pd.read_csv('coffee_survey.csv')

df.head(5)
```

```python
# Show the number of rows and columns of the Data.

df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# See if there are any duplicates in this data set

df.duplicated(keep='last')
```

```python
count_nulls = df.isnull().sum().sort_values(ascending=False)

print(count_nulls)
```

```python
# Replace nulls will 'No Answer'

df2 = df.fillna('No Answer')

df2.head(5)
```

```python
df2.describe()
```

```python
count_nulls2 = df2.isnull().sum().sort_values(ascending=False)

print(count_nulls2)
```

```python
df2 = df2.apply(lambda x: x.str.lower())
df2.head(5)
```

```python
# Export clean data to new csv for analysis

df2.to_csv('coffee_data_clean.csv', sep=',', )
```
